### Remapeamento de features

```python
df.columns
```

```python
df[['event', 'Transplant', 'removed_list']].value_counts()
```

```python
df['time'].describe()
```

```python
import math

id = lambda x: x

imperialize_disease = lambda x: {
        'Outras': 'other',
        'HAS': 'hypertension',
        'Diabetes': 'diabetes',
        'DNC': 'virulent_newcastle_disease',
        'Pielonefrite': 'pyelonephritis',
        'GNC': 'glomerulonephritis'
    }[x]

imperialize_allele_category = lambda x: 'homozygous' if x == 'homozigoto' else 'heterozygous'

translate_boolean = lambda x: x == "Sim"

categorize_event = lambda x: ['waiting', 'transplanted', 'died_waiting', 'removed'][x]

feature_map = {
    "date_registered": ("date", pd.to_datetime),
    "age_registered": ("age_at_list_registration", id),
    "dialysis_session_count": ("time_on_Dialysis", np.vectorize(lambda x: 0 if math.isnan(x) else x)),
    "sex": ("sex", id),
    "race": ("race", id),
    "underlying_disease": ("underline_disease", np.vectorize(imperialize_disease)),
    "diabetes": ("diabetes", lambda x: x == 1),
    "chagas": ("chagas", lambda x: x == 1),
    "blood_type": ("Blood_type", id),
    "transfusion_count": ("number_transfusion", id),
    "gestation": ("gestation", translate_boolean),
    "prior_transplant": ("prior_transplant", translate_boolean),
    "c_pra": ("cPRA", id),
    "hla_a1": ("HLA_A1", id),
    "hla_a2": ("HLA_A2", id),
    "hla_b1": ("HLA_B1", id),
    "hla_b2": ("HLA_B2", id),
    "hla_dr1": ("HLA_DR1", id),
    "hla_dr2": ("HLA_DR2", id),
    "dr_00": ("DR_00", np.vectorize(imperialize_allele_category)),
    "b_00": ("B_00", np.vectorize(imperialize_allele_category)),
    "a_00": ("A_00", np.vectorize(imperialize_allele_category)),
    "anti_hbc": ("anti.HBc", translate_boolean),
    "anti_hcv": ("anti.HCV", translate_boolean),
    "hbs_ag": ("agHBs", translate_boolean),
    "event": ("event", np.vectorize(categorize_event)),
    "days_waiting": ("time", id)
}

df_mapped = pd.DataFrame()
for key, (source, transform) in feature_map.items():
    df_mapped[key] = transform(df[source])
df_mapped.head()
```
